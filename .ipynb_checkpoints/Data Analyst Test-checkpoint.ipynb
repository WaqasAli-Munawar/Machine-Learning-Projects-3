```python
import pandas as pd
import re
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
pd.set_option('display.float_format', lambda x: '%.3f' % x)
sns.set()
pd.options.display.max_rows = None
pd.options.display.max_columns = None
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset = pd.read_excel("GL.XLSX")
```

```python
dataset.columns
```

```python
boolean = dataset["Posting Date"].str.contains(r"[a-z]+")
summary = dataset[boolean == True]["Posting Date"].unique()
summary
```

```python
indx_header = dataset[boolean.isnull()].index
dataset.drop(indx_header, inplace=True)
dataset = dataset.reset_index(drop=True)
```

```python
dataset['Posting Date']=dataset['Posting Date'].apply(str)
```

```python
lst = []

for i in dataset["Posting Date"]:
    if re.search("[a-z]+",i):
        value = i
        lst.append(i)
    else:
        lst.append(value)

dataset["type"] = pd.Series(lst)
```

```python
boolean = dataset["Posting Date"].str.contains(r"[a-z]+")
new_head_index = dataset[boolean].index
```

```python
# new_head_index
```

```python
# dataset.iloc[22165]
```

```python
dataset["head_name"] = "empty"

strt = 0
max_indx = new_head_index.max()

for i in new_head_index[1:]: 
    end = i
    head = dataset.loc[strt,["Posting Date","Due Date"]].astype(str).str.cat(sep=" ")
    dataset.iloc[strt:end,-1] = head
#     print(strt,end)
    strt = end
    
    if strt == max_indx:
        dataset.iloc[strt:,-1] = dataset.loc[strt,["Posting Date","Due Date"]].astype(str).str.cat(sep=" ")
        dataset.iloc[strt:,-1] = head
    
    
    
# dataset[["Posting Date","type","head_name"]]
```

```python
new_boolean = boolean == False
```

```python
# dataset["str_date"] = ""

date_index = dataset.loc[new_boolean,"Posting Date"].index
```

```python
dataset.loc[date_index,"Posting Date"] = pd.to_datetime(dataset.loc[date_index,"Posting Date"])
dataset.loc[date_index,"str_date"] = dataset.loc[date_index,"Posting Date"].apply(lambda x: x.strftime('%b-%Y'))
```

```python
types_sum = \
dataset.pivot_table(index = "type", values = "Cumulative Balance (LC)", aggfunc = "sum", margins = True)


print("Task 1 Summary Table: ")
types_sum
```

```python
x = np.arange(types_sum.shape[0]-1)
height = types_sum["Cumulative Balance (LC)"].values[:-1]

fig, ax = plt.subplots(figsize=(14,7))
plt.bar(x,height, color = "yellow")
plt.xticks(x,types_sum.index[:-1],rotation=90)
ax.ticklabel_format(style='plain', axis='y')
loc = 0
for i in types_sum["Cumulative Balance (LC)"].values[:-1]:
    ax.annotate(s=round(i+loc),xy=(loc,-1000000), rotation = 90, color = "black") # -1000000
    loc += 1

plt.title("Task 1 Summary Graph",{'fontsize': "15",
        'fontweight': "10",
        'color': "black"})
plt.show()
```

```python
type_group = dataset.groupby("type")
```

```python
type_group.size()
```

```python
cost_of_sales = type_group.get_group("Cost of Sales")
```

```python
cost_of_sales_analysis = cost_of_sales.pivot_table(index = "str_date", values = "Cumulative Balance (LC)")\
                                .loc[["Jan-2019","Feb-2019","Mar-2019","Apr-2019","May-2019",\
                                                         "Jun-2019","Jul-2019","Aug-2019","Sep-2019","Oct-2019",\
                                                        "Nov-2019","Dec-2019"]]
print("Cost of Sale Month Wise Summary")
cost_of_sales_analysis
```

```python
x = np.arange(cost_of_sales_analysis.shape[0])
height = cost_of_sales_analysis["Cumulative Balance (LC)"].values

fig, ax = plt.subplots(figsize=(14,7))
plt.bar(x,height, color = "yellow")
plt.xticks(x,cost_of_sales_analysis.index,rotation=90)
ax.ticklabel_format(style='plain', axis='y')
loc = 0
for i in cost_of_sales_analysis["Cumulative Balance (LC)"].values:
    ax.annotate(s=round(i+loc),xy=(loc,500000), rotation = 90, color = "black") # -1000000
    loc += 1

plt.title("Cost of Sale Chart Month Wise",{'fontsize': "15",
        'fontweight': "10",
        'color': "black"})
plt.show()
```

```python
cost_of_sales_head_analysis = cost_of_sales.pivot_table(index = "head_name", values = "Cumulative Balance (LC)")\
                                
print("Cost of Sale Head Wise Summary")
cost_of_sales_head_analysis
```
